```python
import joblib
import gradio as gr
```

```python
model = joblib.load('/content/logistic_regression(sn_and_attd)')
```

```python
def predict(attd1, attd2, attd3, attd4, sn1, sn2, sn3, sn4):
  x = [[attd1, attd2, attd3, attd4, sn1, sn2, sn3, sn4]]
  y = model.predict(x)

  if y[0] == 0:
    y = "Negative Purchase Intention."
  else:
    y = "Positive Purchase Intention"
  return y
```

```python
# Create Gradio interface
interface = gr.Interface(
    fn=predict,
    inputs=
     [  gr.Slider(value=3, minimum=1, maximum=5, step=1, label = "Advertisements on social media can help me to be aware of the existence of a particular product."),
        gr.Slider(value=3, minimum=1, maximum=5, step=1, label ="Advertisements on social media attract my attention more easily than that of other advertising channels, such as, television, etc."),
        gr.Slider(value=3, minimum=1, maximum=5, step=1, label = " Prominent keywords, such as, 'promotion' and 'discount' on social media attract my attention to seek more information about a product."),
        gr.Slider(value=3, minimum=1, maximum=5, step=1, label = "I have purchased products advertised on social media."),
        gr.Slider(value=3, minimum=1, maximum=5, step=1, label = "My family has influence over my purchasing habits through social media marketing."),
        gr.Slider(value=3, minimum=1, maximum=5, step=1, label = "People around me think that I should purchase products through social media."),
        gr.Slider(value=3, minimum=1, maximum=5, step=1, label = "I feel good if many people purchased products through social media."),
        gr.Slider(value=3, minimum=1, maximum=5, step=1, label = "My friends encourage me to purchase products through social media.")
        ],
    outputs=[gr.Textbox(label="Output", lines=3)],
)

# Launch the interface
interface.launch()
```
